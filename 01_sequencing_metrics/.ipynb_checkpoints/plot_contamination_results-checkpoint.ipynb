## The purpose of this script is to combine the results files from the contamination check and plot the results

```r
library(ggplot2)
```

```r
contamination_dir="/lustre/scratch119/humgen/projects/psoriasis/contamination_check/"
metaData <- read.table("/nfs/users/nfs_s/so11/phd/psoriasis/bsub_jupyter_lab/psoriasis/sample_info/sample_meta.txt", h=T)


## Retain only sequenced genomes
metaData <- subset(metaData, Seq_status=="WES")

## Combine the output files into a single file
system(paste("cat ", contamination_dir, "*selfSM | awk 'BEGIN {print \"Sample Contamination\"} $1~/P.*/ {print $1, $7}'  > ", contamination_dir, "contamination_report.txt", sep=""))


## Read in and merge the data
contam_rep <- read.table(paste(contamination_dir, "contamination_report.txt", sep=""), h=T)
metaData <- merge(metaData, contam_rep, by.x="sampleID", by.y="Sample")
```

```r
## Plot the results
png(filename = paste(contamination_dir, "contamination_results.png", sep=""), width=2000, height = 500)
plot <- ggplot(metaData, aes(x=sampleID, y=Contamination, fill=patient_ID)) + geom_bar(stat='identity') +
  scale_y_continuous(labels=scales::percent) + labs(x="") + 
  theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 6))
print(plot)
dev.off()
```

```r
png(filename = paste(contamination_dir, "contamination_results.png", sep=""), width=2000, height = 500)
plot <- ggplot(metaData, aes(x=reorder(sampleID, Contamination), y=Contamination*100)) + geom_point() + 
labs(x="Sample", y="Percentage contamination") + theme_classic(base_size = 40) + theme(axis.text.x = element_blank(), axis.ticks.x=element_blank())
print(plot)
dev.off()


ggplot(metaData, aes(x=reorder(sampleID, Contamination), y=Contamination*100)) + geom_point() + 
labs(x="Sample", y="Percentage contamination") + theme_classic(base_size = 40) + theme(axis.text.x = element_blank(), axis.ticks.x=element_blank())
```
